```python
from imutils import paths
```

```python
imagePaths = list(paths.list_images('dataset'))
```

```python
total = 0
real_count = 0
fake_count = 0
for imagePath in imagePaths:
    total += 1
    if 'real' in imagePath:
        real_count += 1
    elif 'fake' in imagePath:
        fake_count += 1
print('total ', total)
print('real ', real_count)
print('fake ', fake_count)
```
