```python
from imutils import paths
import os
from os.path import join
```

```python
imagePaths = list(paths.list_images('CASIA-SURF'))
```

```python
total = 0
removed = 0
for imagePath in imagePaths:
    total += 1
    if 'depth' in imagePath or 'ir' in imagePath:
        os.remove(imagePath)
        removed += 1

print('total ', total)
print('removed ', removed)
```
